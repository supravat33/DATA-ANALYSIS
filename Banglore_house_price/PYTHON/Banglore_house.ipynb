```python
# Convert ranges like "1200-1500" to average, then make numeric
def convert_sqft(x):
    try:
        if '-' in str(x):
            vals = str(x).split('-')
            return (float(vals[0]) + float(vals[1])) / 2
        return float(x)
    except:
        return None

df['total_sqft'] = df['total_sqft'].apply(convert_sqft)
```

```python
df['total_sqft'].unique()
```

```python
# Extract only the number from size
df['bhk'] = df['size'].str.extract(r'(\d+)').astype(int)
```

```python
# Remove rows where bath > bhk + 2
df = df[df['bath'] < df['bhk'] + 2]
```

```python
# Plot 1: BHK Distributio
plt.figure(figsize=(7,7))
df['bhk'].value_counts().sort_index().plot(kind='bar', color='steelblue', edgecolor='black')
plt.title('BHK Distribution')
plt.xlabel('BHK')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('1_bhk_distribution.png')
plt.close()
```

```python
# Plot 2: Price Distribution
plt.figure(figsize=(8, 5))
df[df['price'] < 500]['price'].hist(bins=50, color='coral', edgecolor='black')
plt.title('Price Distribution (below 500 Lakhs)')
plt.xlabel('Price (Lakhs)')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('2_price_distribution.png')
plt.close()
print("Saved: 2_price_distribution.png")
```
